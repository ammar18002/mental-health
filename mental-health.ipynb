# Cleaned Sleep Duration column

```python
import re

invalid_values = [
    'Work_Study_Hours',
    'Pune',
    '45-48 hours',
    '49 hours',
    '35-36 hours',
    'No',
    '45',
    '55-66 hours',
    'Moderate',
    'Unhealthy',
    '40-45 hours',
    'Sleep_Duration'
]

# Normalize: lowercase + strip + standardize unicode dash
train_data['Sleep Duration'] = (
    train_data['Sleep Duration']
        .astype(str)
        .str.strip()
        .str.replace('–', '-', regex=False)   # convert en-dash to normal dash
        .str.lower()
)

# Build regex pattern to match ANY invalid value
pattern = '|'.join([re.escape(v.lower()) for v in invalid_values])

# Replace anything matching pattern with NaN
train_data.loc[
    train_data['Sleep Duration'].str.contains(pattern, na=False),
    'Sleep Duration'
] = np.nan
```

# Cleaned Degree column

```python
train_data['Degree'].unique()
```

```python
valid_patterns = [ 'BHM', 'LLB', 'B.Pharm', 'BBA', 'MCA', 'MD', 'BSc', 'ME', 'B.Arch',
                  'BCA', 'BE', 'MA', 'B.Ed', 'B.Com', 'MBA', 'M.Com', 'MHM', 'BA',
                  'Class 12', 'M.Tech', 'PhD', 'M.Ed', 'MSc', 'B.Tech', 'LLM',
                  'MBBS', 'M.Pharm', 'UX/UI Designer', 'BEd', 'MPA', 'BH',
                  'B.Sc', 'S.Tech', 'Class 11', 'H_Pharm', 'P.Com', 'BPharm', 'Business Analyst', 'M.Arch',
                  'LL.Com', 'Data Scientist', 'MPharm', 'L.Ed', 'P.Pharm', 'BArch',
                  'S.Pharm', 'LLBA', 'LLCom', 'MTech', 'BPA', 'MEd', 'B BA', 'B.B.Arch',
                  'ACA', 'M_Tech', 'LLTech', 'LHM', 'Lata', 'S.Arch',
                  'B.Student', 'LL B.Ed', 'M.S', 'nan', 'K.Ed',
                  'LLS', 'LLEd', 'E.Tech', 'Doctor', 'N.Pharm', 'LCA',
                  'B B.Com', 'RCA', 'Advait'

]

# Step 1 — Normalize column
train_data['Degree'] = (
    train_data['Degree']
        .astype(str)
        .str.strip()
        .str.replace('–', '-', regex=False)
)

# Step 2 — Build a single big regex pattern

# Escape regex characters and join them safely

valid_regex = r'(' + '|'.join([re.escape(v) for v in valid_patterns]) + r')'

# Step 3 — Mark ALL rows that DO NOT match any valid pattern as NaN
train_data.loc[
    ~train_data['Degree'].str.contains(valid_regex, case=False, na=False), 'Degree'] = np.nan
```

# LowerCasing all strings

```python
cols_to_clean = [
    'Name',
    'Gender',
    'City',
    'Working Professional or Student',
    'Profession',
    'Dietary Habits',
    'Family History of Mental Illness',
    'Have you ever had suicidal thoughts ?'
]

for col in cols_to_clean:
    train_data[col] = (
        train_data[col]
            .str.strip()            # remove extra spaces
            .str.replace('–', '-', regex=False)   # normalize dash
            .str.lower()            # convert to lowercase
    )
```
