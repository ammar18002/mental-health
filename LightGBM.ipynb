```python
import pandas as pd
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_data = pd.read_csv('/Users/abdul/Downloads/mental-health/playground-series-s4e11/train_data_cleaned.csv')
test_data = pd.read_csv('/Users/abdul/Downloads/mental-health/playground-series-s4e11/test_data_cleaned.csv')
```

```python
label = 'Depression'

# LightGBM hyperparameters (safe for your Mac)
hyperparams = {
    'GBM': {                 # LightGBM
        'num_boost_round': 300,
        'learning_rate': 0.04,
        'num_leaves': 64,
        'bagging_fraction': 0.8,
        'feature_fraction': 0.8,
        'early_stopping_rounds': 30,
        'verbosity': 1,
        'n_jobs': 1,     # ← ONLY THIS fixes macOS crashes

    }
}

# Train
predictor = TabularPredictor(
    label=label,
    eval_metric='accuracy'
).fit(
    train_data=train_data,
    hyperparameters=hyperparams,
    time_limit=900,        # 15 minutes max
    auto_stack=False,      # NO stacking → faster + stable
    ag_args_fit={'num_cpus': 2}
)
```
